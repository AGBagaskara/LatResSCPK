Pemerintah daerah sedang mengevaluasi dua pabrik raksasa (Pabrik A dan Pabrik B) terkait regulasi
emisi. Ditemukan sistem persamaan linier dengan x untuk emisi harian Pabrik A, dan y untuk Pabrik B:

3x + 2y = 120

x + 4y = 140

Gunakan fungsi aljabar linier NumPy untuk mencari nilai x dan y, lalu cetak kesimpulan dengan contoh
format : "Emisi Pabrik A adalah ... ton dan Pabrik B adalah ... ton".

```python
kiri = np.array([
    [3,2],
    [1,4]
])

kanan = np.array([120,140])

hasil = np.linalg.solve(kiri, kanan)
namaPabrik = ['Pabrik A', 'Pabrik B']

for i in range(len(hasil)):
    print(f'{namaPabrik[i]} adalah {int(hasil[i])} ton')
```

Buatlah suatu fungsi untuk mengotomatiskan perhitungan emisi dengan spesifikasi :

• Masukan fungsi : Matriks emisi

• Keluaran fungsi 1 : total keseluruhan emisi dari seluruh sektor, dan rata-rata emisi bulanan
untuk setiap sektor (Beri nama fungsi ini dengan : hitung_total_rata2.)

• Keluaran fungsi 2 : nilai emisi tertinggi (maksimal) dan nilai emisi terendah (minimal) dari
seluruh data (Beri nama fungsi ini dengan : cari_ekstrem_emisi)

• Panggil kedua fungsi tersebut dengan memasukkan emisi_matriks

```python
def hitung_total_rata2(matriks_emisi):
    total = np.sum(matriks_emisi, axis=1)
    rata_rata = np.mean(matriks_emisi, axis=1)
    return total, rata_rata
    
def cari_ekstrem_emisi(matriks_emisi):
    matriksReshape = matriks_emisi.flatten()
    dataTertinggi = matriksReshape[np.argmax(emisi_matriks)]
    dataTerendah = matriksReshape[np.argmin(emisi_matriks)]
    return dataTertinggi, dataTerendah

total_emisi, rata_rata_emisi = hitung_total_rata2(emisi_matriks)
data_tertinggi_emisi, data_terendah_emisi = cari_ekstrem_emisi(emisi_matriks)

print(f"Total Emisi          : {total_emisi}")
print(f"Rata2 Emisi          : {rata_rata_emisi}")
print(f"Data Tertinggi Emisi : {data_tertinggi_emisi}")
print(f"Data Terendah Emisi  : {data_terendah_emisi}")
```
