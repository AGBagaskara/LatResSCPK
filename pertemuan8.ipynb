Penentuan **Alternatif** dan **Kriteria**
- Alternatif:
  - A1: Yamaha
  - A2: Honda
  - A3: Suzuki
  - A4: Kawasaki
- Kriteria:
  - C1: Gaya
  - C2: Keandalan
  - C3: Keekonomisan Bahan Bakar

**Tingkat Kepentingan**:
| Intensitas | Keterangan |
|------------|------------|
| 1 | X dan Y sama penting/baik |
| 2 | X sedikit lebih penting/baik dari Y |
| 3 | X lebih penting/baik dari Y |
| 4 | X jauh lebih penting/baik dari Y |

**Preferensi Kriteria**
1. *Keandalan* dua kali lebih penting dari gaya.
2. *Gaya* tiga kali lebih penting dari keekonomisan bahan bakar.
3. *Keandalan* empat kali lebih penting dari keekonomisan bahan bakar.

**Preferensi Alternatif**
1. Preferensi Gaya:
   - Kawasaki > Honda > Yamaha > Suzuki
2. Preferensi Keandalan:
   - Honda > Yamaha > Kawasaki > Suzuki
3. Nilai Keekonomisan Bahan Bakar:
   - Yamaha  : 60
   - Honda   : 80
   - Suzuki  : 60
   - Kawasaki: 80

```python
def normalize_comparation(m): #normalisasi matriks
    sumCol = m.sum(axis=0)
    norm = m/sumCol
    return norm

def weight(m): #menghitung bobot
    return np.mean(m, axis=1)

def validity_check(m, w):
    n = len(m)
    
    RI = {
        2: 0.00, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45,
        10: 1.51, 11: 1.53, 12: 1.54, 13: 1.56, 14: 1.57
    }
    
    CV = m @ w / w
    print(f"CV: ",(CV),"\n")
    
    eigen = np.mean(CV)
    print(f"Eigen: ", (eigen), "\n")
    
    print(f"CI: ",(RI[n]),"\n")
    
    CI = (eigen - n)/(n-1)
    print(f"CI: ",(CI),"\n")
    
    CR = CI/RI[n]
    print(f"CR: ",(CR),"\n")
    
    if CR <= 0.1:
        print("Bobot Valid")
    else:
        print("Tidak Valid")
    
        
def final_weight(w_alt, w_crit):
    w_fin = w_alt.T @ w_crit
    return w_fin
```

### Evaluasi dan Bobot Kriteria

#### Matriks Perbandingan Berpasangan Kriteria

```python
A = ["Yamaha", "Honda", "Suzuki", "Kawasaki"]

print("==== Kriteria ====")

MPBK = np.array([
    [1  , 1/2, 3 ],
    [2  , 1  , 4 ],
    [1/3, 1/4, 1 ]
    ])

print(MPBK)
```

```python
MPBK_norm = normalize_comparation(MPBK)
print(MPBK_norm)

wk = weight(MPBK_norm)
print(wk)

kCheck = validity_check(MPBK, wk)
print(kCheck)
```

### Kriteria Keandalan

```python
MPBa = np.array([
    [1, 1/2,	3,	2],
    [2,	1,	4,	3],
    [1/3, 1/4, 1, 1/2],
    [1/2, 1/3, 2, 1]
])

MPBa_norm = normalize_comparation(MPBa)
print(MPBa_norm)

wa = weight(MPBa_norm)
print(wa)

aCheck = validity_check(MPBa, wa)
print(aCheck)
```

### Kriteria Gaya

```python
MPBb = np.array([
    [1, 1/2,	2,	1/3],
    [2,	1,	3,	1/2],
    [1/2, 1/3, 1, 1/4],
    [3, 2, 4, 1]
])

MPBb_norm = normalize_comparation(MPBb)
print(MPBb_norm)

wb = weight(MPBb_norm)
print(wb)

bCheck = validity_check(MPBb, wb)
print(bCheck)
```

### Kriteria Ekonomis

```python
MPBc = np.array([
    [1, 60/80,	1,	60/80],
    [80/60,	1,	80/60,	1],
    [1, 60/80, 1, 60/80],
    [80/60, 1, 80/60, 1]
])

MPBc_norm = normalize_comparation(MPBc)
print(MPBc_norm)

wc = weight(MPBc_norm)
print(wc)

cCheck = validity_check(MPBc, wc)
print(cCheck)
```

```python
w_total = np.array([wb,wa,wc])
print(w_total)

w_final = final_weight(w_total, wk)
print(w_final)

max_index = np.argmax(w_final)
print("Alternatif terbaik adalah ", A[max_index])
```
