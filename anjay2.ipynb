# Pertemuan 2 ~ Operasi MatrikS

### Deklarasi

```python
import numpy as np 

#vektor baris (1 x n)
vector = np.array([1,2,3])
print("vektor baris : ")
print(vector)

#2d array matriks 
matrix2d = np.array([[1,2,3], [4,5,6]])
print("matriks 2d : ")
print(matrix2d)

#array multidimensi
mdarray = np.array([[9,8,7], [6,5,4], [3,2,1]])
print("matriks multidimensi : ")
print(mdarray)

arrayA = np.array([[3,2,1], [6,5,4], [9,8,7]])

#penjumlahan
pnjmlh = arrayA + mdarray
print("penjumlahan matriks : ")
print(pnjmlh)

#pengurangan
pngrn = arrayA - mdarray
print("pengurangan matriks : ")
print(pngrn)

#perkalian
prkln = np.dot(arrayA, mdarray)
print("perkalian matriks : ")
print(prkln)

arrayB = np.array([[3,2], [6,4], [9,8]])

#perkalian
prkln = np.dot(arrayA, matrix2d.T)
print("perkalian matriks ngawur: ")
print(prkln)
```

### Sistem Persamaan Linear

```python
A = np.array([[1, -2], [12, 5]])
B = np.array([32, 7])
print(np.linalg.solve(A,B))
```

### Penggabungan Matriks

```python
print("naik")
print(np.vstack((arrayA, mdarray)))
print("turun")
print(np.hstack((arrayA, arrayB)))
```

### Reshape

```python
print(arrayB.reshape(2,3))
```

```python

print(arrayA.flatten())
```
