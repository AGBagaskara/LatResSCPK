```python
import numpy as np
import matplotlib.pyplot as plt
import os

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

# Matriks Produksi (jumlah unit yang diproduksi per bulan)
p = np.array([[120, 150, 170],
              [100, 130, 160],
              [110, 140, 180]])

# Matriks Biaya Produksi per unit 
c = np.array([[50, 55, 60],
              [48, 52, 58],
              [47, 50, 57]])

# Matriks harga jual per unit
h = np.array([[80, 85, 90],
              [78, 82, 88],
              [75, 80, 85]])
```

```python
def totalBiayaProduksi():
    # 1️. Total biaya produksi per pabrik per bulan
    totalBiayaProduksi = p * c
    print("=== Total Biaya Produksi per Pabrik per Bulan ===\n")
    for i, baris in enumerate(totalBiayaProduksi):
        print(f"Pabrik {i+1}:")
        for j, nilai in enumerate(baris):
            print(f"  Bulan {j+1}: {nilai}")

    # 2 Total biaya produksi per pabrik selama 3 bulan
    totalBiayaPabrik = np.sum(totalBiayaProduksi, axis=1)
    print("\n=== Total Biaya Produksi per Pabrik (3 Bulan) ===\n")
    for i, total in enumerate(totalBiayaPabrik):
        print(f"Pabrik {i+1}: {total}")
```

```python
def totalPendapatan():
    # 1️. Total Pendapatan per pabrik per bulan
    totalPendapatan = p * h
    print("=== Total Pendapatan per Pabrik per Bulan ===\n")
    for i, baris in enumerate(totalPendapatan):
        print(f"Pabrik {i+1}:")
        for j, nilai in enumerate(baris):
            print(f"  Bulan {j+1}: {nilai}")

    # 2 Total Pendapatan per pabrik selama 3 bulan
    totalPendapatanPabrik = np.sum(totalPendapatan, axis=1)
    print("\n=== Total Pendapatan Produksi per Pabrik (3 Bulan) ===\n")
    for i, total in enumerate(totalPendapatanPabrik):
        print(f"Pabrik {i+1}: {total}")
```

```python
def hitungLabaBersih():
    # Total biaya
    totalBiaya = p * c
    # Total pendapatan
    totalPendapatan = p * h
    # Laba bersih (per bulan, per pabrik)
    labaBersih = totalPendapatan - totalBiaya
    return labaBersih
```

```python
def tampilLabaBersih():
    labaBersih = hitungLabaBersih()

    # 1️. Total Laba Bersih per pabrik per bulan
    print("=== Laba Bersih per Pabrik per Bulan ===\n")
    for i, baris in enumerate(labaBersih):
        print(f"Pabrik {i+1}:")
        for j, nilai in enumerate(baris):
            print(f"  Bulan {j+1}: {nilai}")
    totalLaba = np.sum(labaBersih, axis=1)

    # 2. Total Laba Bersih per pabrik selama 3 bulan
    print("\n=== Total Laba Bersih per Pabrik (3 Bulan) ===\n")
    for i, total in enumerate(totalLaba):
        print(f"Pabrik {i+1}: {total}")
```

```python
def labaBersihTertinggi():
    labaBersih = hitungLabaBersih()
    
    # Total laba bersih selama 3 bulan
    totalLabaPerPabrik = np.sum(labaBersih, axis=1)
    print("=== Total Laba Bersih Selama 3 Bulan ===\n")
    for i, total in enumerate(totalLabaPerPabrik):
        print(f"Pabrik {i+1}: {total}")

    # Cari pabrik dengan laba tertinggi
    indeksTertinggi = np.argmax(totalLabaPerPabrik)

    print("\nPabrik dengan laba bersih tertinggi selama 3 bulan adalah:")
    print(f"Pabrik {indeksTertinggi+1} dengan total laba {totalLabaPerPabrik[indeksTertinggi]}")
```

```python
def visualisasiData():
    bulan = [1, 2, 3]
    produksi = p
    biaya = p * c
    laba = (p * h) - (p * c)

    # Membuat Canvas
    plt.figure(figsize=(8,5)) 

    # 1. Grafik Produksi
    for i, data in enumerate(produksi):
        plt.plot(bulan, data,
                 marker='o',
                 linestyle='-',
                 label=f'Pabrik {i+1}')
    plt.title('Tren Produksi', fontsize=14, fontweight='bold')
    plt.xlabel('Bulan')
    plt.ylabel('Jumlah Produksi')
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.legend()

    # 2. Grafik Biaya Produksi
    plt.figure(figsize=(8,5)) 
    for i, data in enumerate(biaya):
        plt.plot(bulan, data,
                 marker='o',
                 linestyle='-',
                 label=f'Pabrik {i+1}')
    plt.title('Tren Biaya Produksi', fontsize=14, fontweight='bold')
    plt.xlabel('Bulan')
    plt.ylabel('Biaya')
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.legend()

    # 3. Grafik Laba Bersih setiap Pabrik
    plt.figure(figsize=(8,5)) 
    for i, data in enumerate(laba):
        plt.plot(bulan, data,
                 marker='o',
                 linestyle='-',
                 label=f'Pabrik {i+1}')
    plt.title('Tren Laba Bersih per Pabrik', fontsize=14, fontweight='bold')
    plt.xlabel('Bulan')
    plt.ylabel('Laba')
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    while True:
        clear_screen()
        print("+------------------- MENU -------------------+")
        print("|1. Total Biaya Produksi                     |")
        print("|2. Total Pendapatan                         |")
        print("|3. Laba Bersih                              |")
        print("|4. Pabrik dengan Laba Bersih Tertinggi      |")
        print("|5. Visualisasi Data                         |")
        print("|6. Keluar                                   |")
        print("+--------------------------------------------+")
        pilihan = input("Pilih menu (1-6): ")
        match pilihan:
            case "1":
                totalBiayaProduksi()
            case "2":
                totalPendapatan()
            case "3":
                tampilLabaBersih()
            case "4":
                labaBersihTertinggi()
            case "5":
                visualisasiData()
            case "6":
                print("Keluar dari program. Terima Kasih")
                break
            case _:
                print("Pilihan tidak valid, silakan coba lagi.\n")
        input("\nTekan Enter untuk melanjutkan...")
```
