<h1> Tugas 2 SCPK IF-G</h1>

<p>Sebuah universitas ingin memberikan penghargaan kepada mahasiswa terbaik pada akhir semester
melalui proses seleksi yang objektif dan terukur. Untuk membantu proses pengambilan keputusan,
pihak kampus menggunakan Sistem Pendukung Keputusan (SPK) dengan metode Weighted Product
(WP).</p>

<p>Penilaian dilakukan berdasarkan beberapa aspek yang dianggap penting dalam menentukan kualitas
mahasiswa, yaitu kemampuan akademik, prestasi non-akademik, keaktifan organisasi, dan kedisiplinan.
Setiap aspek memiliki tingkat kepentingan yang berbeda sehingga diberikan bobot tertentu.</p>

<p>Terdapat 4 kandidat mahasiswa yang masuk tahap seleksi akhir, dengan data penilaian sebagai
berikut:</p>

| Mahasiswa | IPK | Prestasi | Keaktifan Organisasi | Jumlah Pelanggaran |
|-----------|-----|----------|----------------------|--------------------|
|     A     | 3.8 |    85    |          90          |         2          |
|     B     | 3.6 |    95    |          80          |         1          |
|     C     | 3.9 |    80    |          85          |         3          |
|     D     | 3.7 |    88    |          95          |         1          |

<P>Bobot masing-masing kriteria adalah:
• C1 = 35%
• C2 = 30%
• C3 = 25%
• C4 = 10%
</P>

Soal
1. Tentukan nilai vektor S untuk masing-masing alternatif menggunakan metode Weighted
Product.
2. Hitung nilai preferensi V dari setiap mahasiswa.
3. Urutkan seluruh mahasiswa berdasarkan hasil perhitungan metode WP.
4. Tentukan mahasiswa yang paling layak mendapatkan penghargaan mahasiswa terbaik.

```python
import numpy as np
import pandas as pd
```

```python
alternatif = ["A", "B", "C", "D"]
kriteria = ["IPK", "Prestasi", "Keaktifan Organisasi", "Jumlah Pelanggaran"]
bobot_awal = [0.35, 0.30, 0.25, 0.10]

# Matriks keputusan x
# Baris Alternatif, Kolom Kriteria
data = np.array([
    [3.8, 85, 90, 2],
    [3.6, 95, 80, 1],
    [3.9, 80, 85, 3],
    [3.7, 88, 95, 1]
    ])

# Pengkali kriteria cost = -1, benefit = 1
k = [1, 1, 1, -1]

df = pd.DataFrame(data, index=kriteria, columns=alternatif)
print("=== Matriks Keputusan Awal ===")
print(df)
print("\nBobot Awal (Skala 1-5):", bobot_awal)
```

<h1>Normalisasi Nilai Bobot</h1>

```python
m, n = data.shape
norm_bobot = bobot_awal / np.sum(bobot_awal)

print("Bobot Ternormalisasi:", norm_bobot)
```

<h1>Menghitung Nilai Preferensi (Vektor S)</h1>

```python
s = []
  
s = np.prod(data ** (k * norm_bobot), axis=1)
  
s_df = pd.DataFrame(s, index=alternatif, columns=["Vektor S"])
print(s_df)
```

<h1>Normalisasi Vektor S (Vektor V)</h1>

```python
v = s / np.sum(s)

v_df = pd.DataFrame(v, index=alternatif, columns=["Vektor V"])
sort_ranking = v_df.sort_values(by='Vektor V', ascending=False)
print("=== Hasil Ranking ===")
print(sort_ranking)
print()

best_alternative = sort_ranking.index[0]
print(f"Alternatif terbaik adalah {best_alternative}")
```

<p>Berdasarkan perhitungan metode Weighted Product (WP), urutan peringkat mahasiswa dari yang tertinggi ke terendah adalah D, B, A, C, sehingga mahasiswa D merupakan kandidat terbaik untuk mendapatkan penghargaan.</p>
