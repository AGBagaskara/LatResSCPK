# SPK dengan SAW

## 1. Import Library

```python
import numpy as np
```

## 2. Mempersiapkan Data yang Diperlukan

#### 2.1 Inisialisasi Data

```python
x = np.array([[150, 15, 2, 2, 3],
[500, 200, 2, 3, 2],
[200, 10, 3, 1, 3],
[350, 100, 3, 1, 2]])
```

#### 2.2 Menentukan atribut (0=biaya, 1=keuntungan)

```python
k = np.array([0, 1, 1, 0, 1])
```

#### 2.3 Menentukan bobot setiap kriteria

```python
w = np.array([0.25, 0.15, 0.30, 0.25, 0.05])
```

## 3. Proses SPK

#### 3.1 Normalisasi Matriks

```python
m, n = x.shape # Matriks ukuran m x n
R = np.zeros((m, n)) # Membuat matriks R kosong
for j in range(n):
    if k[j] == 1: # Kriteria keuntungan
        R[:, j] = x[:, j] / np.max(x[:, j])
    else: # Kriteria biaya
        R[:, j] = np.min(x[:, j]) / x[:, j]
```

#### 3.2 Perangkingan Matriks

```python
V = np.sum(w * R, axis=1)
```

## 4. Hasil

#### 4.1 Nilai V setiap alternative

```python
V
```

#### 4.2 Melihat alternatif terbaik

```python
best_alternative_index = np.argmax(V)

print(f"Alternatif terbaik adalah A{best_alternative_index + 1} dengan nilai {V[best_alternative_index]:.4f}")
```
