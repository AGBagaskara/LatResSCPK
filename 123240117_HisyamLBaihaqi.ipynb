```python
import numpy as np
import matplotlib.pyplot as plt
import os

def clear_screen():
    # 'cls' untuk Windows, 'clear' untuk Linux/Mac
    os.system('cls' if os.name == 'nt' else 'clear')

P = np.array([[120,150,170],
              [100,130,160],
              [110,140,180]])

C = np.array([[50,55,60],
              [48,52,58],
              [47,50,57]])

H = np.array([[80,85,90],
              [78,82,88],
              [75,80,85]])

def total_biaya_produksi():
    totBiayaProduksi = P * C
    # axis = 1 untuk menjumlahkan secara horizontal
    totBiayaPabrik = np.sum(totBiayaProduksi, axis=1)

    # fungsi enumerate untuk membungkus list tersebut dan memberikan nomor antrean pada setiap datanya jadi bisa dikeluarkan
    # satu persatu
    for i, total in enumerate(totBiayaPabrik):
        print(f"Pabrik {i+1}: {total}")

    nama_pabrik = ['Pabrik 1', 'Pabrik 2', 'Pabrik 3']

    # plt bar untuk membuat grafik dengan pakai matplotlib
    plt.bar(nama_pabrik, totBiayaPabrik, color=['red','blue','pink'])

    # 4. Tambahkan Keterangan
    plt.title('Total Biaya Produksi per Pabrik')
    plt.xlabel('Nama Pabrik')
    plt.ylabel('Total Biaya')

    plt.show()

def total_pendapatan():
    totPerkalianPendapatan = P * H
    # axis = 1 untuk menjumlahkan secara horizontal
    totPendapatanPabrik = np.sum(totPerkalianPendapatan, axis=1)

    # fungsi enumerate untuk membungkus list tersebut dan memberikan nomor antrean pada setiap datanya jadi bisa dikeluarkan
    # satu persatu
    for i, total in enumerate(totPendapatanPabrik):
        print(f"Pabrik {i+1}: {total}")

    nama_pabrik = ['Pabrik 1', 'Pabrik 2', 'Pabrik 3']

    # plt bar untuk membuat grafik dengan pakai matplotlib
    plt.bar(nama_pabrik, totPendapatanPabrik, color=['red','blue','pink'])

    # 4. Tambahkan Keterangan
    plt.title('Total Pendapatan per Pabrik')
    plt.xlabel('Nama Pabrik')
    plt.ylabel('Total Pendapatan')

    plt.show()

def laba_bersih():
    totBiayaProduksi = P * C
    # axis = 1 untuk menjumlahkan secara horizontal
    totBiayaPabrik = np.sum(totBiayaProduksi, axis=1)

    # fungsi enumerate untuk membungkus list tersebut dan memberikan nomor antrean pada setiap datanya jadi bisa dikeluarkan
    # satu persatu
    for i, total in enumerate(totBiayaPabrik):
        print(f"Pabrik {i+1}: {total}")

    nama_pabrik = ['Pabrik 1', 'Pabrik 2', 'Pabrik 3']

    # plt bar untuk membuat grafik dengan pakai matplotlib
    plt.bar(nama_pabrik, totBiayaPabrik, color=['red','blue','pink'])

    # 4. Tambahkan Keterangan
    plt.title('Total Biaya Produksi per Pabrik')
    plt.xlabel('Nama Pabrik')
    plt.ylabel('Total Biaya')

    plt.show()

def total_laba_bersih():
    totPerkalianPendapatan = P * H
    # axis = 1 untuk menjumlahkan secara horizontal
    totPendapatanPabrik = np.sum(totPerkalianPendapatan, axis=1)

    totBiayaProduksi = P * C
    # axis = 1 untuk menjumlahkan secara horizontal
    totBiayaPabrik = np.sum(totBiayaProduksi, axis=1)

    totLabaBersih = totPendapatanPabrik - totBiayaPabrik

    for i, total in enumerate(totLabaBersih):
        print(f"Pabrik {i+1}: {total}")

    #########################################################################################################################

    nama_pabrik = ['Pabrik 1', 'Pabrik 2', 'Pabrik 3']

    # plt bar untuk membuat grafik dengan pakai matplotlib
    plt.bar(nama_pabrik, totLabaBersih, color=['red','blue','pink'])

    # 4. Tambahkan Keterangan
    plt.title('Total Laba Bersih per Pabrik')
    plt.xlabel('Nama Pabrik')
    plt.ylabel('Total Laba Bersih')

    plt.show()

def pabrikLabaTertinggi():
    totPerkalianPendapatan = P * H
    # axis = 1 untuk menjumlahkan secara horizontal
    totPendapatanPabrik = np.sum(totPerkalianPendapatan, axis=1)

    totBiayaProduksi = P * C
    # axis = 1 untuk menjumlahkan secara horizontal
    totBiayaPabrik = np.sum(totBiayaProduksi, axis=1)

    totLabaBersih = totPendapatanPabrik - totBiayaPabrik

    terbesar = 0
    for i in range(len(totLabaBersih)):
        if totLabaBersih[i] > totLabaBersih[terbesar]:
            terbesar = i

    print(f"Pabrik dengan laba tertinggi adalah pabrik : {terbesar+1} \nPabrik tersebut memiliki laba bersih yaitu : {totLabaBersih[terbesar]}")

import numpy as np
import matplotlib.pyplot as plt

# Data dari soal [cite: 5, 8, 11]
P = np.array([[120, 150, 170], [100, 130, 160], [110, 140, 180]]) # Produksi
C = np.array([[50, 55, 60], [48, 52, 58], [47, 50, 57]])         # Biaya
H = np.array([[80, 85, 90], [78, 82, 88], [75, 80, 85]])         # Harga Jual

def menu_visualisasi():
    totalUnitProduksi = np.sum(P, axis=1)

    totPerkalianPendapatan = P * H
    # axis = 1 untuk menjumlahkan secara horizontal
    totPendapatanPabrik = np.sum(totPerkalianPendapatan, axis=1)

    totBiayaProduksi = P * C
    # axis = 1 untuk menjumlahkan secara horizontal
    totBiayaPabrik = np.sum(totBiayaProduksi, axis=1)

    totLabaBersih = totPendapatanPabrik - totBiayaPabrik



    pabrik = ['Pabrik 1', 'Pabrik 2', 'Pabrik 3']
    #untuk membuat koordinat titik pusat karena nanti kan disatukan 3 table atau batang menjadi satu tempat bukan berbeda
    x = np.arange(len(pabrik)) # lokasi [0, 1, 2]
    lebar = 0.25 # lebar tiap batang

    # 2. Membuat Plot
    #membuat batang untuk biaya produksi , x - lebar untuk posisi batand dimana dikurangi lebar batang yang sudah diinisialisasi di awal

    plt.bar(x - lebar/2, totBiayaPabrik, lebar, label='Biaya Produksi', color='red')
    plt.bar(x + lebar/2, totLabaBersih , lebar, label='Laba Bersih', color='pink')
    # Keterangan grafik doang

    # label untuk sumbu x
    plt.xlabel('Daftar Pabrik')
    # label untuk sumbuy
    plt.ylabel('Nilai')
    # judul table
    plt.title('Biaya, dan Laba Bersih')

    # mengganti angka koordinat dari 0 1 2 menjadi pabrik  1 2 3
    plt.xticks(x, pabrik)

    # membuat legenda
    plt.legend()

    # membuat visual enak dilihat mata saja
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # tampilkan

    # eksekusi akhir untuk memunculkan semua layout yang dibutuhkann
    plt.tight_layout()
    plt.show()

    ## membuat plot untuk tren produksi
    plt.subplot(2, 1, 2)
    plt.plot(pabrik, totalUnitProduksi, marker='o', linestyle='-', color='blue', label='Tren Produksi')

    plt.xlabel('Daftar Pabrik')
    plt.ylabel('Total Unit')
    plt.title('Tren Total Produksi per Pabrik')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # eksekusi akhir untuk memunculkan semua layout yang dibutuhkann
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    while True:
        clear_screen()
        print("+------------------- MENU -------------------+")
        print("|1. Total Biaya Produksi                     |")
        print("|2. Total Pendapatan                         |")
        print("|3. Laba Bersih                              |")
        print("|4. Pabrik dengan Laba Tertinggi             |")
        print("|5. Visualisasi Laba Bersih(Opsional)        |")
        print("|6. Keluar                                   |")
        print("+--------------------------------------------+")

        pilihan = input("Pilih menu (1-6): ")

        match pilihan:
            case "1":
                print("Menampilkan Total Biaya Produksi\n")
                total_biaya_produksi()
            case "2":
                print("Menampilkan Total Pendapatan\n")
                total_pendapatan()
            case "3":
                print("Menampilkan Laba Bersih\n")
                total_laba_bersih()
            case "4":
                print("Menampilkan Pabrik dengan Laba Tertinggi\n")
                pabrikLabaTertinggi()
            case "5":
                print("Menampilkan Visualisasi Laba Bersih\n")
                menu_visualisasi()
            case "6":
                print("Keluar dari program. Sampai jumpa!")
                break
            case _:
                print("Pilihan tidak valid, silakan coba lagi.\n")
        input("\nTekan Enter untuk melanjutkan...")
```
